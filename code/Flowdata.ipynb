```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar e preparar dados
df = pd.read_csv('flowdata.csv', sep=';')
df = df.rename(columns={
    'Emitter': 'emitters',
    'FlowMeasured': 'flow_measured',
    'FlowReference': 'flow_reference'
})
df = df.sort_values('emitters')

# Criar figura
plt.figure(figsize=(12, 6))

# Plotar dados
plt.plot(df['emitters'], df['flow_measured'],
         'b^-', label='Flow Measured', markersize=8, linewidth=1.5)
plt.plot(df['emitters'], df['flow_reference'],
         'rv-', label='Flow Reference', markersize=8, linewidth=1.5)

# Formatar eixo X (apenas nmeros pares)
plt.xticks(range(2, 37, 2))
plt.xlim(0.5, 36.5)
plt.xlabel('Emitter Number')

# Formatar eixo Y (sem decimais)
plt.gca().yaxis.set_major_formatter(plt.FormatStrFormatter('%d'))
plt.ylabel('Flow Rate (Lh)')

# Adicionar grade
plt.grid(True, alpha=0.3, linestyle='--')

# Legenda no estilo Excel (acima do grfico, horizontal, sem caixa)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),
           ncol=2, frameon=False, fontsize=11)

# Ajustar layout
plt.tight_layout()

# Nome do arquivo
filename = 'flow_rate_plot.tiff' # Default filename

# Salvar no formato TIFF (PLOS ONE)
plt.savefig(filename,
            format='tiff',
            dpi=300,
            bbox_inches='tight',
            pad_inches=0.1,
            facecolor='white')

plt.show()
```
